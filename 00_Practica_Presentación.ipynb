## ***Métodos de suavizado (SMA,WMA y EMA) Practica Presentación***

imagen.png

```python
import numpy as np
import pandas as pd


# Serie de tiempo
data = [10, 12, 13, 16, 15, 20, 22, 21, 23, 25,
        26, 28, 30, 29, 31, 33, 35, 36, 38, 42]

serie = pd.Series(data)

#1. Media Móvil Simple (SMA)
k = 3
sma = serie.rolling(window=k).mean()


#2. Media Móvil Ponderada (WMA)
weights = np.array([0.2, 0.3, 0.5])

def wma(values):
    return np.dot(values, weights)

wma_series = serie.rolling(window=3).apply(wma, raw=True)


#3. Media Móvil Exponencial (EMA)
alpha = 0.3
ema = serie.ewm(alpha=alpha, adjust=False).mean()

#Resultados
df = pd.DataFrame({
    "y": serie,
    "SMA": sma,
    "WMA": wma_series,
    "EMA": ema
})

print(df)

```
