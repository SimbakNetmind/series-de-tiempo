# Error Absoluto Medio (MAE)

El **MAE** (Mean Absolute Error) es una métrica fundamental para evaluar el desempeño de modelos de regresión.

### 📋 Definición y Características
* **Promedio de magnitudes:** Mide el promedio de las magnitudes absolutas de los errores.
* **Interpretación Directa:** Esta métrica es especialmente valorada por su interpretación directa, ya que se expresa en las **mismas unidades de la variable original**.
* **Robustez:** No penaliza de manera desproporcionada los errores grandes, lo que la hace una medida **robusta frente a valores atípicos** (outliers) moderados.
* **Importancia:** Es apropiada cuando todos los errores tienen la misma importancia.

> **Nota:** No distingue entre errores pequeños y grandes más allá de su magnitud absoluta.

---

### 🔢 Formulación Matemática

La fórmula para calcular el MAE es:

$$\text{MAE} = \frac{1}{T} \sum_{t=1}^{T} |Y_t - \hat{Y}_t|$$

Donde:
* $T$: Número total de observaciones.
* $Y_t$: Valor real.
* $\hat{Y}_t$: Valor predicho por el modelo.

# Requerimientos para la Aplicación de MAE

Para que el cálculo del **Error Absoluto Medio (MAE)** sea válido y proporcione información útil sobre el desempeño de un modelo, se deben cumplir los siguientes requisitos:

### 1. Naturaleza del Problema
* **Tarea de Regresión:** El MAE está diseñado exclusivamente para problemas de **regresión** (predicción de valores numéricos continuos). No es aplicable a tareas de clasificación donde las etiquetas son categóricas.
* **Escala de los Datos:** Las variables deben estar en una escala cuantitativa (intervalo o razón).

### 2. Estructura de los Datos
* **Pares de Comparación:** Se requiere un conjunto de datos que contenga:
    * **Valores Reales ($Y_t$):** Las etiquetas verdaderas u observaciones históricas.
    * **Valores Predichos ($\hat{Y}_t$):** Las estimaciones generadas por el modelo.
* **Mismo Tamaño ($T$):** Ambos vectores (reales y predicciones) deben tener exactamente la misma longitud o número de muestras.

### 3. Consistencia de Unidades
* **Unidades Homogéneas:** Tanto $Y_t$ como $\hat{Y}_t$ deben estar expresados en las **mismas unidades** de medida (por ejemplo, metros, dólares o grados Celsius).
* **Sin Transformaciones Inconsistentes:** Si los datos reales están en escala logarítmica, las predicciones también deben estarlo para que la resta $|Y_t - \hat{Y}_t|$ tenga sentido físico.

### 4. Supuestos sobre los Errores
* **Importancia Lineal:** Se debe asumir que el costo de un error aumenta de forma **lineal**. Es decir, un error de 10 unidades es exactamente el doble de grave que un error de 5 unidades.
* **Presencia de Outliers:** El MAE es ideal cuando el conjunto de datos tiene valores atípicos moderados y no queremos que estos "distorsionen" la métrica final (a diferencia de lo que ocurre con el MSE).

---

### Resumen Matemático de Componentes
Para ejecutar la fórmula $$\text{MAE} = \frac{1}{T} \sum_{t=1}^{T} |Y_t - \hat{Y}_t|$$ necesitas asegurar que:
1. $Y_t, \hat{Y}_t \in \mathbb{R}$
2. $T > 0$ (El conjunto no puede estar vacío).
3. No existan valores nulos (`NaN`) en ninguno de los dos vectores, ya que invalidarían la sumatoria.

## 🎯 ***Practica codigo (MAE)***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear el DataFrame
data = {
    "Dia": np.arange(1, 21),
    "Valor_real": [95, 97, 96, 98, 100, 102, 101, 103, 105, 104,
                   106, 108, 107, 109, 110, 112, 111, 113, 114, 115],
    "Pronostico_A": [92, 93, 98, 94, 96, 99, 97, 105, 100, 101,
                     110, 104, 109, 104, 106, 115, 108, 110, 118, 111],
    "Pronostico_B": [94, 96, 95, 99, 100, 101, 100, 103, 104, 105,
                     105, 108, 106, 110, 109, 112, 112, 114, 113, 115],
    "Pronostico_C": [90, 101, 92, 103, 95, 107, 96, 99, 110, 98,
                     112, 102, 113, 103, 116, 106, 117, 107, 120, 109]
}



df = pd.DataFrame(data)


def mae(y, y_hat):
    return np.mean(np.abs(y - y_hat))

metricas = pd.DataFrame({
    "MAE": [
        mae(df["Valor_real"], df["Pronostico_A"]),
        mae(df["Valor_real"], df["Pronostico_B"]),
        mae(df["Valor_real"], df["Pronostico_C"])
    ]})

metricas
```

### 🎯 Conclusión del Ejemplo MAE

Analizando los resultados de la métrica MAE para los diferentes pronósticos, se observa que:

*   **Pronóstico B** tiene el **MAE más bajo (0.75)**. Esto significa que, en promedio, las predicciones de este modelo están más cerca de los valores reales, con una desviación de 0.75 unidades.
*   **Pronóstico A** presenta un **MAE de 3.50**.
*   **Pronóstico C** tiene el **MAE más alto (5.40)**, lo que indica que sus predicciones son las que más se desvían de los valores reales.

**En resumen, el Pronóstico B es el modelo con mejor rendimiento** en este conjunto de datos, ya que un MAE menor indica una mayor precisión en las predicciones.
